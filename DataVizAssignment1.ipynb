Here's the grouped bar chart showing children's feelings about science before and after the program. This visualization should clearly communicate the positive shift in perceptions due to the pilot program, with a noticeable increase in the proportion of children who are excited about science and a decrease in negative feelings.

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Excited', 'Kind of Interested', 'OK', 'Not Great', 'Bored']
pre_values = [0.19, 0.25, 0.40, 0.11, 0.05]
post_values = [0.38, 0.30, 0.14, 0.12, 0.06]

n_categories = len(categories)

index = np.arange(n_categories)

bar_width = 0.35

fig, ax = plt.subplots()

pre_bars = ax.bar(index, pre_values, bar_width, label='Pre program', color='blue')
post_bars = ax.bar(index + bar_width, post_values, bar_width, label='Post program', color='red')

ax.set_title('Children\'s Feelings About Science Before and After the Program')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(categories)
ax.set_ylabel('Proportion of Responses')

ax.legend()
ax.yaxis.grid(True)

plt.tight_layout()
plt.show()


```

Here's the stacked 100% bar chart summarizing the survey results. It shows the change in proportions of children's feelings about science before and after the program.

     
Each colored segment represents the percentage of responses for each feeling, showing a clear shift in sentiment from 'Pre' to 'Post'.

```python
# Creating a stacked 100% bar chart to summarize the survey results

# The proportion values for 'Pre' and 'Post' need to be converted to percentages
# and stacked in such a way that each bar sums up to 100%.
pre_percentages = [value * 100 for value in pre_values]
post_percentages = [value * 100 for value in post_values]

categories = ['Pre program', 'Post program']
cumulative_pre = np.cumsum(pre_percentages)
cumulative_post = np.cumsum(post_percentages)

fig, ax = plt.subplots()

base_pre = np.zeros(1)
base_post = np.zeros(1)

colors = ['blue', 'lightblue', 'red', 'pink', 'orange']

for i in range(len(pre_percentages)):
    ax.bar(categories[0], pre_percentages[i], bottom=base_pre, color=colors[i], edgecolor='white', width=0.4)
    base_pre += pre_percentages[i]

for i in range(len(post_percentages)):
    ax.bar(categories[1], post_percentages[i], bottom=base_post, color=colors[i], edgecolor='white', width=0.4)
    base_post += post_percentages[i]

ax.set_title('Comparison of Children\'s Feelings About Science Before and After the Program')
ax.set_ylabel('Percentage of Responses')
ax.legend(categories, loc='upper left', bbox_to_anchor=(1,1), title='Survey')

plt.tight_layout()
plt.show()
```
